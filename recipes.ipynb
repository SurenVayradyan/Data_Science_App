```python
n_ingred_train = X_train.sum(axis=1).to_frame(name='n_ingred')
n_ingred_test = X_test.sum(axis=1).to_frame(name='n_ingred')
```

```python
reg = LinearRegression().fit(n_ingred_train, y_train)
```

```python
y_pred_sum = reg.predict(n_ingred_test)
mean_squared_error(y_test, y_pred_sum, squared=False)
```

```python
n_ingred_train_scaled = (n_ingred_train - n_ingred_train.mean())/n_ingred_train.std()
```

### Sparse least squares
